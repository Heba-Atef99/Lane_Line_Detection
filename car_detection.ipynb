# 6.Run The Model

Define the path to the folder on drive with all needed files

````python
path = '/content/drive/MyDrive/car_detection/'
v1 = 'project_video.mp4'
v2 = 'challenge_video.mp4'
v3 = 'harder_challenge_video.mp4'

output1 = 'result1'
output2 = 'result2'
output3 = 'result3'
````

Detect the cars for chosen video

````python
## VIDEO DETECTION 
detector_video = VideoObjectDetection()
# Choose YOLO v3 Tiny
detector_video.setModelTypeAsTinyYOLOv3()
# Load weights
detector_video.setModelPath(path + 'yolo-tiny.h5')
detector_video.loadModel()
# Make model detect cars only
custom = detector_video.CustomObjects(car=True)

# print(detector_video)
video_path = detector_video.detectCustomObjectsFromVideo(custom_objects=custom, 
                                                         input_file_path=path + v3,
                                                         output_file_path=path + output3,
                                                         frames_per_second=20, log_progress=False)
print(video_path)
````
